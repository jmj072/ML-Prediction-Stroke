# **Stroke Prediction**


## 데이터 분석 배경 

**Behaviral RskBehavioral Risk Factor Surveillance System  (BRFSS) 2015 Dataset**  
[BRFSS 2015 Code book](https://www.cdc.gov/brfss/annual_data/2015/pdf/codebook15_llcp.pdf)


1998년에서 2009년까지의 통계청 자료에 의하면, 뇌졸중(Stroke)은 암 다음으로 흔한 사망원인으로, 단일 장기질환 중 사망률 1위의 심각한 질환입니다.  
또한 뇌졸중은 발생 후에 치료를 하더라도 후유증이나 합병증으로 인하여 반신마비, 언어장애, 심하면 사망에 이를 수 있는 매우 위험한 병입니다.

각 개인들이 가지고 있는 질환과 생활 패턴을 통해 뇌졸중에 걸릴 확률을 미리 예측하여 예방할 수 있다면,  
뇌졸중으로 인해 발생하는 가정적, 사회적 손실을 방지할 수 있을 것입니다.

BRFSS 2015 Data Set은 미국 주 별로 성인들의 만성 질환 여부, 흡연, 신체활동, 과일 야채 섭취 등 다양한 생활 패턴을 전화조사(telephon surveys)한 데이터 입니다.  

이 데이터를 활용하여 뇌졸중 발병 확률에 영향을 주는 feature 들을 확인해보고, 각 개인들의 뇌졸중 발병 확률을 예측해 보겠습니다.


뇌졸중의 위험인자(risk factor)로 알려진 것은 다음과 같습니다.

1. 조절가능한 위험인자
    - 고혈압
    - 심장질환
    - 당뇨병
    - 흡연
    - 과음 ( More than 2 drinks per day )
    - 식이 (포화 지방, 트랜스 지방 및 콜레스테롤이 많은 식단, 나트륨 함량이 많거나 칼로리가 많은 식단.  하루에 5인분 이상의 과일과 채소를 포함하는 식단은 뇌졸중 위험을 줄일 수 있음)
    - 고지혈증
    - 운동부족
    - 비만
    - 스트레스
    - 사회적/경제적 요인 (소득이 낮은 사람들)

2. 조절불가능한 위험인자
    - 연령 (55세 이후에는 10년마다 두 배씩 유병확률이 증가함.)
    - 인종 (African Americans)
    - 성별 (남성이 뇌졸중에 걸릴 확률이 높으나, 사망률은 여성이 더 높음.)

## **변수 선택 및 설명**

* Dependent Variables  
    - **Stroke** : (Ever told) you had a stroke.

* Independent Variables
    - 고혈압
        >**HighBP** (Adults who have been told they have high blood pressure by a doctor, nurse, or other health professional)  
    > *BPMEDS (Are you currently taking medicine for your high blood pressure?)*  나중에 고려해야 할수도 있음.
    - 고지혈증
    > **HighChol**  (Have you EVER been told by a doctor, nurse or other health professional that your blood cholesterol is high?)  
    > *_CHOLCHK (Cholesterol check within past five years)* 나중에 고려해야 할수도 있음.
    - 당뇨
        > **Diabetes** ((Ever told) you have diabetes (If "Yes" and respondent is female, ask "Was this only when you were pregnant?". If Respondent says pre-diabetes or borderline diabetes, use response code 4.) )  
    > *INSULIN (Are you now taking insulin?)* 나중에 고려해야 할수도 있음.
    - 심장질환
        >**HeartDisease** (Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI))
    - 흡연
    > **Smoker** (Four-level smoker status: Everyday smoker, Someday smoker, Former smoker, Non-smoker)
    - 음주
    > **HvyAlcohol** (Heavy drinkers : adult men having more than 14 drinks per week and adult women having more than 7 drinks per week)
    - 식이
    > **Fruits** (Consume Fruit 1 or more times per day)  
    > **Veggies** (Consume Vegetables 1 or more times per day)
    - 운동부족
    > **PhysActivity** (Adults who reported doing physical activity or exercise during the past 30 days other than their regular job)  
    > **DiffWalk** (Do you have serious difficulty walking or climbing stairs?)
    - 비만
    > **BMIGroup** (Four-categories of Body Mass Index (BMI), Underweight, Normal Weight, Overweight, Obese)
    - 스트레스
    > **MentHlth** (Now thinking about your mental health, which includes stress, depression, and problems with emotions, for how many days during the past 30 days was your mental health not good?)

    - 소득
    > **Income** ( Is your annual household income from all sources: (If respondent refuses at any income level, code "Refused."))
    - 학력
    >**Education** (What is the highest grade or year of school you completed?)
    - 혼인 여부
    > **Marital** (Are you: (marital status))
    - 의료보험 여부
    > **AnyHealthcare**(Do you have any kind of health care coverage, including health insurance, prepaid plans such as HMOs, or government plans such as Medicare, or Indian Health Service?)
    - 의료비용 소비 여부
    > **NoDocbcCost** (Was there a time in the past 12 months when you needed to see a doctor but could not because of cost?

    - 나이
    > AgeGroup

    - 인종
    > Race

    - 성별
    > Sex

## Dataset 나누기

```python
train, test = train_test_split(df, test_size=0.2, random_state=2)
train, val = train_test_split(train, test_size=0.2, random_state=2)

train.shape, val.shape, test.shape
```

```python
target = 'Stroke'
features = df.columns.drop(target)

X_train = train[features]
y_train = train[target]

X_val = val[features]
y_val = val[target]

X_test = test[features]
y_test = test[target]
```

## Base Model 설정 
**Logistic Regression**

```python
pipe_Logistic = make_pipeline(
    OrdinalEncoder(), 
    LogisticRegression(class_weight = 'balanced' , random_state=2, solver='liblinear')
)

pipe_Logistic.fit(X_train, y_train)
y_pred = pipe_Logistic.predict(X_val)
y_pred_proba = pipe_Logistic.predict_proba(X_val)[:,-1]
print(classification_report(y_val, y_pred))
print('AUC score: ', roc_auc_score(y_val, y_pred_proba))
```

```python
model_lr = pipe_Logistic.named_steps['logisticregression']
column_names = X_val.columns
coefficients = pd.Series(model_lr.coef_[0], column_names)
plt.figure(figsize=(10,10))
coefficients.sort_values().plot.barh();
```

```python
sns.set_style('white')
fig, ax = plt.subplots(figsize = (5,5))
pcm = plot_confusion_matrix(pipe_Logistic, X_val, y_val,
                        values_format = 'd',
                        display_labels = ['Normal', 'Stroke'],
                        ax = ax);
```

## Decision Tree Model

### F1 score

```python
pipe_Decision = make_pipeline(
    OrdinalEncoder(), 
    DecisionTreeClassifier(class_weight = 'balanced' , random_state=2)
)

# 튜닝할 하이퍼파라미터의 범위를 지정해 주는 부분
dists = {
    'decisiontreeclassifier__criterion' : ['gini', 'entropy'],
    'decisiontreeclassifier__max_depth': randint(1,21),
    'decisiontreeclassifier__min_samples_split':  uniform(0,1),
    'decisiontreeclassifier__min_samples_leaf': randint(1,11),
    'decisiontreeclassifier__max_features': ['sqrt', 'log2', 1, 5, 10, 15, 20]
}

clf_Decision = RandomizedSearchCV(
    pipe_Decision, 
    param_distributions=dists, 
    n_iter=50, 
    cv=5, 
    scoring='f1' ,
    verbose=1,
    n_jobs=-1
)

clf_Decision.fit(X_train, y_train)
pd.DataFrame(clf_Decision.cv_results_).sort_values(by='rank_test_score').T
```

```python
pipe_Decision_Best = clf_Decision.best_estimator_
y_pred = pipe_Decision_Best.predict(X_val)
print(classification_report(y_val,y_pred))
```

## SHAP

```python
explainer = shap.TreeExplainer(Boosting_best)
row = X_test.iloc[[1]]
row_encoded = encoder.transform(row)
shap_values = explainer.shap_values(row_encoded)

shap.initjs()
shap.force_plot(
    base_value=explainer.expected_value, 
    shap_values=shap_values, 
    features=row, 
    link='logit' # SHAP value를 확률로 변환해 표시합니다.
)
```

```python
def explain(row_number):
    positive_class = 'Stroke'
    positive_class_index = 1

    # row 값을 변환합니다
    row = X_test.iloc[[row_number]]
    row_processed = encoder.transform(row)

    # 예측하고 예측확률을 얻습니다 
    pred = Boosting_best.predict(row_processed)[0]
    pred_proba = Boosting_best.predict_proba(row_processed)[0, positive_class_index]
    pred_proba *= 100
    if pred != positive_class:
        pred_proba = 100 - pred_proba

    # 예측결과와 확률값을 얻습니다
    print(f'이 환자에 대한 예측결과는 {pred} 으로, 확률은 {100-pred_proba:.0f}% 입니다.')
    
    # SHAP를 추가합니다
    shap_values = explainer.shap_values(row_processed)

    # Fully Paid에 대한 top 3 pros, cons를 얻습니다
    feature_names = row.columns
    feature_values = row.values[0]
    shaps = pd.Series(shap_values[0], zip(feature_names, feature_values))
    pros = shaps.sort_values(ascending=False)[:3].index
    cons = shaps.sort_values(ascending=True)[:3].index

    # 예측에 가장 영향을 준 top3
    print('\n')
    print('Positive 영향을 가장 많이 주는 3가지 요인 입니다:')
    
    evidence = pros if pred == positive_class else cons
    for i, info in enumerate(evidence, start=1):
        feature_name, feature_value = info
        print(f'{i}. {feature_name} : {feature_value}')

    # 예측에 가장 반대적인 영향을 준 요인 top1
    print('\n')
    print('Negative 영향을 가장 많이 주는 3가지 요인 입니다:')
    
    evidence = cons if pred == positive_class else pros
    for i, info in enumerate(evidence, start=1):
        feature_name, feature_value = info
        print(f'{i}. {feature_name} : {feature_value}')

    # SHAP
    shap.initjs()
    return shap.force_plot(
        base_value=explainer.expected_value, 
        shap_values=shap_values, 
        features=row, 
        link='logit'
    )
```
